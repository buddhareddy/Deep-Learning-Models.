```python
# text preporcessing stopwords removal and stemming
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
import nltk
import re # regurlar expression
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
wordnet=WordNetLemmatizer()
```

```python
# just an example for stemming and lemmatization
text='Benoît Hamon Wins French Socialist Party’s Presidential Nomination - The New York Times'
text=text.split()
print(text)
review=[ps.stem(word) for word in text if not word in stopwords.words('english')]
print(review)
review1=[wordnet.lemmatize(word) for word in text if not word in stopwords.words('english')]
print(review1)
```

```python
corpus=[]
for i in range(0,len(finaldf)):
  review=re.sub('[^A-Za-z]', ' ',finaldf['title'][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus[6]
```

```python
# Apply count vectorizer
cv=CountVectorizer(max_features=5000,ngram_range=(1,3))
X=cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
y=finaldf['label']
```

```python
X
```

```python
# to see the featurenames
cv.get_feature_names_out()[:20]
```

```python
#to see the parameters
cv.get_params()
```

```python
#perfrom traina nd text split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.33,random_state=0)
```

```python
count_df=pd.DataFrame(X_train,columns=cv.get_feature_names_out())
count_df.head()
```

```python
#multinomial navie bayes
from sklearn.naive_bayes import MultinomialNB
classfier=MultinomialNB()
```

```python
from sklearn import metrics
import numpy as np
import itertools
```

```python
classfier.fit(X_train,y_train)
y_pred=classfier.predict(X_test)
score=metrics.accuracy_score(y_test,y_pred)
score
```

```python
# confusion matrix
print(metrics.confusion_matrix(y_test,y_pred))
```
