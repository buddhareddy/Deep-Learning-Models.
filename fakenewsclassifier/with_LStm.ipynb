```python

# import libraries
import pandas as pd
```

```python

# load the data
df=pd.read_csv('/content/drive/MyDrive/fakenews/train.csv')
```

```python

#the number of null values
df.isnull().sum().sum()
```

```python
# we have to remove the null values
df=df.dropna()
df.shape
```

```python

#importing libraries related to text to vector conversion
from sklearn.feature_extraction.text import CountVectorizer
```

```python

finaldf=df.copy()
```

```python
finaldf.reset_index(inplace=True)
```

```python
# text preporcessing stopwords removal and stemming
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python

import nltk
import re # regurlar expression
nltk.download('stopwords')
nltk.download('wordnet')
```

```python

# just an example for stemming and lemmatization
text='Benoît Hamon Wins French Socialist Party’s Presidential Nomination - The New York Times'
text=text.split()
#print(text)
review=[ps.stem(word) for word in text if not word in stopwords.words('english')]
#print(review)
```

```python

corpus=[]
for i in range(0,len(finaldf)):
  review=re.sub('[^A-Za-z]', ' ',finaldf['title'][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python

# Apply count vectorizer
cv=CountVectorizer()
X=cv.fit_transform(corpus).toarray()
```

```python

y=finaldf['label']
```

```python

#perfrom traina nd text split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.33,random_state=0)
```

```python

count_df=pd.DataFrame(X_train,columns=cv.get_feature_names_out())
count_df.head()
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size1=5000
```

```python
onehot_repr=[one_hot(words,voc_size1) for words in corpus]
```

```python
onehot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense
```

```python
# emebeded represnetation
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
```

```python
embedded_docs[0]
```

```python
embedding_features=40
model=Sequential()
model.add(Embedding(voc_size1,embedding_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs)
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_final,y,test_size=.3,random_state=1)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=(model.predict(x_test) > 0.5).astype("int32")
```

```python
predict_x=model.predict(x_test)
classes_x=np.argmax(predict_x,axis=1)
```
